```python
import cv2
pathStr = "/kaggle/input/image-colorization-dataset/data/test_color/image5000.jpg"
img = cv2.imread(pathStr)
img.shape
```

```python
def process_img(imgPath, saveName):
    img = cv2.imread(imgPath)
    img = cv2.resize(img, (256,256))
    print("Image has been resized to " + str(img.shape))
    cv2.imwrite(saveName, img)
```

```python
import os
def convert():
    counter = 0
    pathStr = "/kaggle/input/image-colorization-dataset/data/test_color"
    for filename in os.listdir(pathStr):
        counter += 1
        f = os.path.join(pathStr, filename)
        # checking if it is a file
        if os.path.isfile(f):
            saveName = "Result_Image_"+str(counter)+".png"
            print(str(f))
            process_img(f, saveName)
        if(counter > 50):
            break
```

```python
convert()
```
